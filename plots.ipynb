```python
from dataclasses import asdict
from datetime import datetime
from typing import Any

from src.utils.training_logger import load_training_log, list_training_runs, TrainingLog
from src.plotting import gradient_boosting, dn_length_prediction


def _fmt_scalar(v: Any) -> str:
    if isinstance(v, float):
        return f"{v:.4e}"
    if isinstance(v, datetime):
        return v.isoformat(sep=" ", timespec="seconds")
    return str(v)


def _print_pretty(obj: Any, indent: int = 0) -> None:
    """Print dicts and lists with increasing indentation; scalars on one line."""
    pad = "  " * indent
    if isinstance(obj, dict):
        for k, v in obj.items():
            if isinstance(v, dict):
                if v:
                    print(f"{pad}{k}:")
                    _print_pretty(v, indent + 1)
                else:
                    print(f"{pad}{k}: {{}}")
            elif isinstance(v, list):
                if v:
                    print(f"{pad}{k}:")
                    _print_pretty(v, indent + 1)
                else:
                    print(f"{pad}{k}: []")
            else:
                print(f"{pad}{k}: {_fmt_scalar(v)}")
    elif isinstance(obj, list):
        for i, item in enumerate(obj):
            if isinstance(item, dict):
                if item:
                    print(f"{pad}[{i}]")
                    _print_pretty(item, indent + 1)
                else:
                    print(f"{pad}[{i}]: {{}}")
            elif isinstance(item, list):
                if item:
                    print(f"{pad}[{i}]")
                    _print_pretty(item, indent + 1)
                else:
                    print(f"{pad}[{i}]: []")
            else:
                print(f"{pad}[{i}] {_fmt_scalar(item)}")
    else:
        print(f"{pad}{_fmt_scalar(obj)}")


def print_log(log: TrainingLog) -> None:
    _print_pretty(asdict(log), 0)
```
